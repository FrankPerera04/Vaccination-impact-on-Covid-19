```python
main_df = pd.read_csv('/content/drive/MyDrive/Colab Data Files /Data Visualisation Portfolio /owid-covid-data.csv')
```

```python
main_df
```

```python
# Filtering the dataset
```

```python
filtered_df = main_df[main_df['location'].isin(['United States'])]
```

```python
relavant_columns = [
    'location',
    'date',
    'total_cases',
    'new_cases',
    'total_deaths',
    'new_deaths',
    'new_cases_per_million',
    'total_vaccinations',
    'new_vaccinations',
    'people_vaccinated_per_hundred'
]

filtered_df = filtered_df[relavant_columns]
```

```python
# Removing Null Values
```

```python
filtered_df = filtered_df.dropna()
```

# **Exploratory Data Analysis**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
import matplotlib.ticker as ticker
import matplotlib.dates as mdates

# Style for Seaborn
sns.set(style="whitegrid")

df = filtered_df
```

### **Univariate Analysis**

```python
# Distribution of Total Cases

# Get comma separators
def format_with_commas(x, pos):
    return f'{int(x):,}'

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="total_cases", kde=True, bins=30, color='teal', edgecolor='black')
plt.xlabel('Total Cases', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Distribution of Total Cases', fontsize=16)

# Customize x axis
ax = plt.gca()
ax.xaxis.set_major_formatter(FuncFormatter(format_with_commas))
ax.xaxis.set_tick_params(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# New Cases, and Vaccinations Over Time

# Get thousand separator
def format_with_commas(x, pos):
    return f'{int(x):,}'

df['date'] = pd.to_datetime(df['date'])

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="date", y="new_cases", label="New Cases", color="blue")
sns.lineplot(data=df, x="date", y="new_vaccinations", label="New Vaccinations", color="green")

#Customize X and Y axis
ax = plt.gca()
ax.yaxis.set_major_formatter(FuncFormatter(format_with_commas))
plt.xticks(rotation=45)

plt.title("New Cases, and Vaccinations Over Time", fontsize=14)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Total Cases and New Cases over time

# Get comma seperator
def format_with_commas(x, pos):
    return f'{int(x):,}'

plt.figure(figsize=(15, 5))

# Plot 1: Total Cases Over Time
plt.subplot(1, 2, 1)
plt.plot(df['date'], df['total_cases'], label='Total Cases')
plt.title('Total Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
# Customize x and y axis
ax1 = plt.gca()
ax1.yaxis.set_major_formatter(FuncFormatter(format_with_commas))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

# Plot 2: New Cases Over Time
plt.subplot(1, 2, 2)
plt.plot(df['date'], df['new_cases'], label='New Cases')
plt.title('New Cases Over Time')
plt.xlabel('Date')
plt.ylabel('New Cases')
# Customize x and y axis
ax2 = plt.gca()
ax2.yaxis.set_major_formatter(FuncFormatter(format_with_commas))
ax2.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

### **Multivariate Analysis**

```python
# Correlation Matrix

numeric_df = df.select_dtypes(include=['number'])
cor_matrix = numeric_df.corr()

plt.figure(figsize=(14, 12))
sns.set(font_scale=1.2)
heatmap = sns.heatmap(cor_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1,
                      annot_kws={"size": 12},
                      cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Correlation Matrix Heatmap', fontsize=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Trends Over Time: Cases, Deaths, and Vaccinations
df['date'] = pd.to_datetime(df['date'])
plt.figure(figsize=(12, 6))

# Plotting the data
sns.lineplot(data=df, x="date", y="new_cases", label="New Cases", color="blue")
sns.lineplot(data=df, x="date", y="new_deaths", label="New Deaths", color="red")
sns.lineplot(data=df, x="date", y="new_vaccinations", label="New Vaccinations", color="green")
plt.title("Trends Over Time: Cases, Deaths, and Vaccinations")
plt.xlabel("Date")
plt.ylabel("Count")

# Customize y and x axis
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))
```
